 # 3. OPENING A CSV FILE FROM AN URL 

```python
#  THROUGH THE SERVER SE KOI FILE 
```

- requests.get() → downloads the raw CSV data from the URL.

- StringIO → acts like a file object, allowing pandas to read from the response text.

- pd.read_csv() → loads the CSV into a DataFrame.

- headers → mimic a browser request (sometimes needed for protected URLs).

```python
import requests
import pandas as pd
from io import StringIO
# Example CSV URL
url = "https://people.sc.fsu.edu/~jburkardt/data/csv/hw_200.csv"
headers = {
    "User-Agent": "Mozilla/5.0"
}
# Fetch the data
response = requests.get(url, headers=headers)
# Convert response text into a StringIO object
csv_data = StringIO(response.text)
# Read CSV with pandas
df = pd.read_csv(csv_data)
df
```
