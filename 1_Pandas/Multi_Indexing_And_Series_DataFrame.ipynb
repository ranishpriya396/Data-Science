### Wide Format :
- wide format is where , we have a single row for every datapoint with multiple columns to hold the values of various attributes .

### Long Format : 
- In a long format (also called tidy or stacked data), each row represents a single observation or datapoint. Instead of spreading different variables across multiple columns, you compress them into key-value pairs across multiple rows.

```python
import numpy as np 
import pandas as pd 
```

```python
index = [(12, 'esc'),(13,'ecs'),(14,'ecs'), (15,'cse'), (16,'cse'),(17,'cse')]
```

```python
name = ['Ranish', "Gargi","Samradhi","Puja","ritu","Jigyasa"]
```

```python
pd.Series(name , index = index)
```

```python
multi_index = pd.MultiIndex.from_tuples(index)
```

```python
multi_index.levels[0]
```

```python
s1 = pd.Series(name , index = multi_index)
```

```python
m_i = pd.MultiIndex.from_product([['cse', 'ecs'],[12 , 13 ,14 ]] )
```

```python
m_i
```

```python
s2 = pd.Series(name , index = m_i)
```

```python
s2
```

```python
series_to_df = s2.unstack()
```

```python
series_to_df
```

```python
df_to_series = series_to_df.stack()
```

```python
df_to_series
```
