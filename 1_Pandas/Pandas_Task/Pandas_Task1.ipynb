### 1.write a function to change the datatypes of a gien column of a series . Function take a series as input , returned the converted series .
- series = pd.Series([1,2,"python",2.0,True,100]) to float type data

```python
import pandas as pd
```

```python
series = pd.Series([1,2,"python",2.0,True,100])
```

```python
def changing_datatype(series):
   return  pd.to_numeric(series, errors= "coerce")
```

```python
changing_datatype(series)
```

```python
item.dropna()
```

```python
def converting_into_rupees(x):
    try :
        y = x[1:]
    except:
        y = x
    return float(y)*94.33
```

```python
item['item_price'] = item['item_price'].apply(converting_into_rupees)
```

```python
item.sample()
```

```python
# find the mean price:
item['item_price'].mean()
```

```python
# fill missing avlues with mean:
item['item_price'].fillna(item['item_price'].mean())
```
