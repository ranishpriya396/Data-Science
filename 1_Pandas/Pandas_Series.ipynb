```python
import numpy as np 
import pandas as pd 
```

```python
# Series using list :
country = ["India", "Russia","China","Itlay"]
indexing = [1,2,3,4]
s1 = pd.Series(country,index = indexing, name = "country")
```

```python
s1
```

```python
#  Series using dictionary :
c_dict = {"India" : 1 , "Russia" : 2, "China" : 3 , "Italy":4}
s2 = pd.Series(c_dict, name = "country details" ) 
s2
```

```python
#  Series Attributes :: 
print(s1.index)
print(s1.size)
print(s1.dtype)
print(s1.name)
print(s1.is_unique)
print(s1.values)
print(s1.astype)
```

```python
# series methods
s1.head()
s1.tail()
s1.sample()
s1.value_counts()  #-> count the values how many times repeated
s1.sort_index(ascending = False, inplace = True)
s1.info()
```
