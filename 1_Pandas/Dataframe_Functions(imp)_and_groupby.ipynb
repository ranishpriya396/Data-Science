```python
#  Find the ranking of each of the movie in the group according to the IMBD rating :

def movie_ranking(group):
    group["genre_ranking"] = group['IMDB_Rating'].rank(ascending = False)
    return group
```

```python
genres.apply(movie_ranking)
```

```python
movie1.columns
```

```python
#  find the normalized  IMBD rating groupwise :
def normalized(group):
    group['normalized_ranking'] = (group['IMDB_Rating'] -  group['IMDB_Rating'].min() )/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group
```

```python
genres.apply(normalized)
```

```python
# create a function that can return the highest score of any batsman in the any match :
tempdf = ipl1[ipl1.batsman == 'V Kohli']
tempdf.groupby('match_id')['batsman_runs'].sum().sort_values(ascending = False).head().values[0]
```

```python
def highest(batsman):
    tempdf = ipl1[ipl1.batsman == batsman]
    return tempdf.groupby('match_id')['batsman_runs'].sum().sort_values(ascending = False).head().values[0]
    
```
